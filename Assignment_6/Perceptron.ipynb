For this assignment, we will be implementing a Perceptron model using different approaches, analyzing their results. 

## Approaches
### Approach 1: Heuristic Approach
### Approach 2: Gradient Descent Approach

## Setup

```python
import math
import random
from dataclasses import dataclass
from typing import List, Tuple

import matplotlib.pyplot as plt

# Import data.
# If pandas not installed, uncomment as needed:
#!pip -q install pandas
import numpy as np
import pandas as pd
data_file = pd.read_csv("data-1.csv", header=None)
p_data = data_file[[0,1,2]].values

# Make a plot to show data.
plt.figure(figsize=(6, 6))
# Scatter Plot
plt.scatter(p_data[:, 0][p_data[:, 2] == 0], p_data[:, 1][p_data[:, 2] == 0], color='blue', label='Type 0')
plt.scatter(p_data[:, 0][p_data[:, 2] == 1], p_data[:, 1][p_data[:, 2] == 1], color='orange', label='Type 1')
# Show
plt.show()

# Set learning rate to be used.
learning_rate = 0.1
epoch_count = 1000
```

## Approach 1: Heuristic Approach

```python
# Plot setup.
plt.figure(figsize=(6, 6))
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("Solution boundary\n" + "(Learning Rate: " + str(learning_rate) + ")" + "(Iteration: " + str(epoch_count) + ")")
# Scatter Plot
plt.scatter(p_data[:, 0][p_data[:, 2] == 0], p_data[:, 1][p_data[:, 2] == 0], color='blue', label='Type 0')
plt.scatter(p_data[:, 0][p_data[:, 2] == 1], p_data[:, 1][p_data[:, 2] == 1], color='orange', label='Type 1')

np.random.seed(1)

weights = np.random.randn(2)
bias = np.random.randn()

result_history = []

for i in range(epoch_count):
    result_history.append((weights.copy(), bias))
    for x in range(len(p_data)):
        x_0 = p_data[x][0]
        x_1 = p_data[x][1]
        y_true = p_data[x][2]
        
        y_hat = 1 if weights[0] * x_0 + weights[1] * x_1 + bias >= 0 else 0
        
        if y_hat != y_true:
            if y_hat == 0:
                bias = bias + learning_rate
                weights[0] = weights[0] + (learning_rate * x_0)
                weights[1] = weights[1] + (learning_rate * x_1)
            else:
                bias = bias - learning_rate
                weights[0] = weights[0] - (learning_rate * x_0)
                weights[1] = weights[1] - (learning_rate * x_1)
    # Print results
    #print(weights, " ", bias)

result_history.append((weights.copy(), bias))

# Plot lines
x_vals = np.linspace(0, 1, 100)
# Initial (red)
w, b = result_history[0]
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, color='red')
# Progression (green)
for w, b in result_history[1:-1]:
    y_vals = (-w[0] * x_vals - b) / w[1]
    plt.plot(x_vals, y_vals, 'g--', alpha=0.5)
#Final (black)
w, b = result_history[-1]
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, color='black', linewidth=2)

plt.show()
```

## Approach 2: Gradient Descent Approach

```python
# Plot setup.
plt.figure(figsize=(6, 6))
plt.xlim(0, 1)
plt.ylim(0, 1)
# Scatter Plot
plt.scatter(p_data[:, 0][p_data[:, 2] == 0], p_data[:, 1][p_data[:, 2] == 0], color='blue', label='Type 0')
plt.scatter(p_data[:, 0][p_data[:, 2] == 1], p_data[:, 1][p_data[:, 2] == 1], color='orange', label='Type 1')

np.random.seed(1)

weights = np.random.randn(2)
bias = np.random.randn()

result_history = []
error_history = []

for i in range(epoch_count):
    result_history.append((weights.copy(), bias))
    total_error = 0.0
    for x in range(len(p_data)):
        x_0 = p_data[x][0]
        x_1 = p_data[x][1]
        y_true = p_data[x][2]
        
        z = weights[0] * x_0 + weights[1] * x_1 + bias
        y_hat = 1 / (1 + np.exp(-z))
        
        y_error = (y_true - y_hat)
        total_error = total_error + (y_error**2)
        
        bias = bias + (learning_rate * y_error)
        weights[0] = weights[0] + (learning_rate * y_error * x_0)
        weights[1] = weights[1] + (learning_rate * y_error * x_1)
    total_error = total_error / len(p_data)
    error_history.append(total_error)
    
    # If error is small enough, break.
    if total_error <= 0.1:
        break
    
result_history.append((weights.copy(), bias))

plt.title("Solution boundary\n" + "(Learning Rate: " + str(learning_rate) + ")" + "(Iteration: " + str(len(result_history)-1) + "/" + str(epoch_count) + ")")

# Plot lines
x_vals = np.linspace(0, 1, 100)
# Initial (red)
w, b = result_history[0]
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, color='red')
# Progression (green)
for w, b in result_history[1:-1]:
    y_vals = (-w[0] * x_vals - b) / w[1]
    plt.plot(x_vals, y_vals, 'g--', alpha=0.5)
#Final (black)
w, b = result_history[-1]
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, color='black', linewidth=2)

plt.show()


# Plot error history.
plt.figure(figsize=(6, 4))
plt.xlim(0, len(error_history))
plt.ylim(0, 1)
plt.plot(range(len(error_history)), error_history, color='blue')

plt.title("Error Plot")
plt.xlabel("Number of epochs")
plt.ylabel("Error")

plt.show()
```
